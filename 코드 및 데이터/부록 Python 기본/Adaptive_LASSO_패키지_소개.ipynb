### Adaptive LASSO 패키지 소개 FRED-MD

- asgl: https://github.com/alvaromc317/asgl
- 자세한 내용은 위 링크
- Adaptive LASSO는 가능하지만, Elastic Net은 없는 상태

```python
import pandas as pd
import numpy as np
import asgl
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNetCV
```

```python
# 데이터 불러오기
FREDMD = pd.read_excel("C:/Users/A/OneDrive/바탕 화면/RA/Machine Learning/Python Code/r_practice/FREDMDdata.xlsx", sheet_name=0)
y = FREDMD['INDPRO'].values
X = FREDMD.drop(columns=['INDPRO'])
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
## First Stage
panalties = np.append(np.logspace(10, 0, 100), 0)
model = ElasticNetCV(l1_ratio=0.5, alphas=panalties, fit_intercept=True).fit(X_scaled, y)
coef = model.coef_

# 가중치 구하기
penalty = (np.abs(coef) + 1 / np.sqrt(len(y)))
```

```python
## Second Stage (asgl 패키지 활용)
# penalization="alasso" (Adaptive LASSO 옵션)
# lasso_weights=penalty (Adaptive LASSO의 가중치 부분)
model = asgl.ASGL(
            model="lm",
            penalization="alasso",
            parallel=False,
            lasso_weights=penalty)
model.fit(x=X_scaled, y=y)
```
